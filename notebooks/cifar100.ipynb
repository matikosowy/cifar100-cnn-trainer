```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
transform = transforms.ToTensor()

train_dataset = torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform)
```

# Plot sample images

```python
def imshow(img):
    img = img.numpy().transpose((1, 2, 0))
    plt.imshow(img)
    plt.axis('off')

fig, axes = plt.subplots(3, 3, figsize=(6, 6))
for i in range(3):
    for j in range(3):
        index = torch.randint(len(train_dataset), (1,)).item()
        img, label = train_dataset[index]
        axes[i, j].imshow(img.permute(1, 2, 0))
        axes[i, j].set_title(train_dataset.classes[label])
        axes[i, j].axis('off')

plt.tight_layout()
plt.show()
```

# Data statistics

```python
num_classes = len(train_dataset.classes)
print(f"Number of Classes: {num_classes}")
```

```python
num_train_images = len(train_dataset)
num_test_images = len(test_dataset)
print(f"Number of Training Images: {num_train_images}")
print(f"Number of Testing Images: {num_test_images}")
```

```python
image_shape = train_dataset[0][0].shape
print(f"Image Dimensions (Channels, Height, Width): {image_shape}")
```

```python
num_samples = sum(1 for _, label in train_dataset if label == 0)
print(f"Number of samples per class: {num_samples}")
```
